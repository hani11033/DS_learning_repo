## Tokenization

Example: Word and sentence tokenization

```
# import required packages
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Welcome to the course! We are excited to have you here. Today we will learn about NLP."

# tokenize into words
words = word_tokenize(text)
print("Word tokens: " + str(words))

# tokenize into sentences
sents = sent_tokenize(text)
print("Sentence tokens: " + str(sents))
```

## Stop Words

image.png

Example: Removing stop words

```
# import required packages
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# example text
text = "We are practicing removing stop words in a text"

# tokenize into words
tokens = word_tokenize(text)
print("All tokens: " + str(tokens)) 

# remove stop words
stops = stopwords.words()
tokens_no_stops = [t for t in tokens if t not in stops]
print("No stop tokens: " + str(tokens_no_stops))

# customize stop words
custom_stops = stopwords.words()
custom_stops.append("We")

# remove custom stop words
tokens_no_custom_stops = [t for t in tokens if t not in custom_stops]
print("No custom stop tokens: " + str(tokens_no_custom_stops))
```

## Alphanumerics

image.png

Example: Filtering character and numeric tokens

```
# import necessary packages
from nltk.tokenize import word_tokenize

# example text
text = "Intro2NLP is being taught on July 25!"

# tokenize text
tokens = word_tokenize(text)
print("All tokens: " + str(tokens))

# keep only character tokens
tokens_alpha = [t for t in tokens if t.isalpha()]
print("Character tokens: " + str(tokens_alpha))

# keep only numeric tokens
tokens_digit = [t for t in tokens if t.isdigit()]
print("Numeric tokens: " + str(tokens_digit))

# keep character and numeric tokens
tokens_alnum = [t for t in tokens if t.isalnum()]
print("Alphamumeric tokens: " + str(tokens_alnum))
```

## Lowercasing

image.png

Example: Lowercasing text

```
# import required packages
from nltk.tokenize import word_tokenize

# example text
text = "I go to school in College Station, TX"

# tokenize text
tokens = word_tokenize(text)
print("Original tokens: " + str(tokens))

# lowercase tokens
tokens_lower = [t.lower() for t in tokens]
print("Lowercase tokens: " + str(tokens_lower))
```

## Stemming

image.png

Example: Stemming text

```
# import required packages
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# example text
text = "test tests cacti cactus universe university"

# tokenize the text
tokens = word_tokenize(text)
print("All tokens: " + str(tokens))

# instantiate the stemmer
porter = PorterStemmer()

# stem the tokens
tokens_stemmed = [porter.stem(t) for t in tokens] 
print("Stemmed tokens: " + str(tokens_stemmed))
```

## Lemmatization

image.png

Example: Lemmatizing text

```
# import required packages
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# example text
text = "test tests cacti cactus universe university"

# tokenize the text
tokens = word_tokenize(text)
print("All tokens: " + str(tokens))

# instantiate the stemmer
lemmatizer = WordNetLemmatizer()

# stem the tokens
tokens_lemmatized = [lemmatizer.lemmatize(t) for t in tokens] 
print("Lemmatized tokens: " + str(tokens_lemmatized))
```

## Part of Speech Tags

## Named Entity Recognition

image.png

Example: Named entity recognition

```
# import required packages
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.chunk import ne_chunk

# example text
text = "The Eiffel Tower is in Paris"

# tokenize text
tokens = word_tokenize(text)

# tag tokens to part of speech
tokens_pos = pos_tag(tokens)

# create named entity chunks
tokens_chunked = ne_chunk(tokens_pos, binary=True)

# extract named entities
for chunk in tokens_chunked:
      if hasattr(chunk, 'label'):
         print(chunk.label(), ' '.join(c[0] for c in chunk))
```

## Basic Features

image.png

Example: Find number of words

```
# import required packages
from nltk.tokenize import word_tokenize

# create text example
text = ["hullabaloo canek canek","hullabaloo canek canek", "goodbye to texas university","so long to the orange and white"]

# tokenize text
tokens = [word_tokenize(sent) for sent in text]

# calculate number of words
num_words = [len(sent) for sent in tokens]
print(num_words)
```

Example: Find average length of words

```
# import required packages
from nltk.tokenize import word_tokenize

# create text example
text = ["hullabaloo canek canek","hullabaloo canek canek", "goodbye to texas university","so long to the orange and white"]

# tokenize text
tokens = [word_tokenize(sent) for sent in text]

# calculate length of words
word_lengths = [[len(word) for word in sent] for sent in tokens]

# calculate average word length
avg_lengths = [round(sum(sent) / len(sent), 2) for sent in word_lengths]
print(avg_lengths)
```

```
# import required packages
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# create text
text = ["the cat sat", "the cat in the hat", "the cat sat in the hat"]
text_series = pd.Series(text)

# build vectorizer
vect = CountVectorizer(stop_words='english', ngram_range=(1,2))

# fit vectorizer and create bag of words
X_matrix = vect.fit_transform(text_series)

# create data frame
X_df = pd.DataFrame(X_matrix.toarray(), columns=vect.get_feature_names())
print(X_df.head())
```

## Word Embedding

image.png

image.png

Example: Create Word2Vec vector

```
# import required packages
from gensim.models import Word2Vec
from nltk.corpus import gutenberg, stopwords

# load text example (comes tokenized)
text = gutenberg.words(fileids='carroll-alice.txt')

# preprocess text
tokens = [word.lower() for word in text if word.isalpha()]

# create model
model = Word2Vec([tokens], min_count=3, window=3)

# compute pairwise similarity
print(model.wv.similarity('alice','her'))

# find most similar words
print(model.wv.most_similar('read'))
```

## Logistic Regression

image.png

```
import pandas as pd
df = pd.read_csv('IMDB Dataset.csv', nrows=5000)
```

Example: BoW and Logistic Regression

```
X = df.review
y = df.sentiment

# import required packages
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# build vectorizer
vect = CountVectorizer(stop_words='english', ngram_range=(1,2))

# fit and transform training data
X_train_bow = vect.fit_transform(X_train)

# transform testing data
X_test_bow = vect.transform(X_test)

# train logistic regression model
clf = LogisticRegression(max_iter=2000).fit(X_train_bow, y_train)

# predict labels
preds = clf.predict(X_test_bow)

# calculatre accuracy
score = clf.score(X_test_bow, y_test)
print(score)

# calculate confusion matrix
cm = confusion_matrix(y_test, preds) 
print(cm / len(y_test))
```

## Naive Bayes

image.png

Example: Tfidf and Naive Bayes

```
X = df.review
y = df.sentiment

# import required packages
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# build vectorizer
vect = TfidfVectorizer(stop_words='english', ngram_range=(1,2))

# fit and transform training data
X_train_bow = vect.fit_transform(X_train)

# transform testing data
X_test_bow = vect.transform(X_test)

# create classifier
clf = MultinomialNB()

# fit classifier
clf.fit(X_train_bow, y_train)

# make predictions
preds = clf.predict(X_test_bow)

# calculatre accuracy
score = clf.score(X_test_bow, y_test)
print(score)

# calculate confusion matrix
cm = confusion_matrix(y_test, preds) 
print(cm / len(y_test))
```
